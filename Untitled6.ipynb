```python
#EDA 1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime


df = pd.read_csv("BIKE DETAILS.csv")
```

```python
df
```

```python
# 1. What is the range of selling prices in the dataset?
print("Selling Price Range:", df['selling_price'].min(), "to", df['selling_price'].max())
```

```python
#2. What is the median selling price for bikes in the dataset?
print("Median Selling Price:", df['selling_price'].median())
```

```python
#3.  What is the most common seller type?
print("Most Common Seller Type:", df['seller_type'].mode()[0])
```

```python
#4. How many bikes have driven more than 50,000 kilometers?
print("Bikes Driven > 50,000 km:", df[df['km_driven'] > 50000].shape[0])
```

```python
#5. What is the average km_driven value for each ownership type?
print("\nAverage km_driven by ownership type:\n", df.groupby('owner')['km_driven'].mean())
```

```python
#6. What proportion of bikes are from the year 2015 or older.
prop_older_2015 = (df[df['year'] <= 2015].shape[0]) / df.shape[0]
print(f"\nProportion of bikes from 2015 or older: {prop_older_2015:.2%}")
```

```python
#7.  What is the trend of missing values across the dataset?
print("\nMissing Values:\n", df.isnull().sum())
```

```python
#8.  What is the highest ex_showroom_price recorded, and for which bike.
max_price = df['ex_showroom_price'].max()
bike_name = df[df['ex_showroom_price'] == max_price]['name'].values[0]
print("\nMost Expensive Bike:", bike_name, "| Price:", max_price)
```

```python
#9. What is the total number of bikes listed by each seller type?
print("\nTotal Bikes by Seller Type:\n", df['seller_type'].value_counts())
```

```python
# 10. What is the relationship between selling_price and km_driven for first-owner bikes.
sns.scatterplot(data=df[df['owner'] == '1st owner'], x='km_driven', y='selling_price')
plt.title('Selling Price vs KM Driven (1st Owner)')
plt.xlabel('KM Driven')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()
```

```python
#11.  Identify and remove outliers in the km_driven column using the IQR method?
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df[(df['km_driven'] >= lower) & (df['km_driven'] <= upper)]
print("Removed outliers:", df.shape[0] - df_clean.shape[0])
```

```python
# 12. Perform a bivariate analysis to visualize the relationship between year and selling_price?
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='year', y='selling_price')
plt.xticks(rotation=45)
plt.title("Selling Price by Year")
plt.tight_layout()
plt.show()
```

```python
# 13.  What is the average depreciation in selling price based on the bike's age (current year - manufacturing year).
df['age'] = datetime.now().year - df['year']
avg_depreciation = df.groupby('age')['selling_price'].mean()
print("Average Depreciation by Age:\n", avg_depreciation)
```

```python
#14. Which bike names are priced significantly above the average price for their manufacturing year.
avg_price_year = df.groupby('year')['selling_price'].mean().reset_index()
merged = df.merge(avg_price_year, on='year', suffixes=('', '_avg'))
above_avg_bikes = merged[merged['selling_price'] > 1.5 * merged['selling_price_avg']]
print("Bikes Priced Significantly Above Average:\n", above_avg_bikes[['name', 'year', 'selling_price']])
```

```python
#15. Develop a correlation matrix for numeric columns and visualize it using a heatmap.
numeric_cols = ['selling_price', 'year', 'km_driven', 'ex_showroom_price']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# EDA-2
df = pd.read_csv("Car Sale.csv")
```

```python
df
```

```python
# 1. What is the average selling price of cars for each dealer, and how does it compare across different dealers.
dealer_avg_price = df.groupby('Dealer_Name')['Price ($)'].mean().sort_values()
dealer_avg_price.plot(kind='bar', title='Average Car Price by Dealer')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()
```

```python
#2.  Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends.
brand_price_std = df.groupby('Company')['Price ($)'].std().sort_values(ascending=False)
print("Brand with highest price variation:", brand_price_std.idxmax())
brand_price_std.plot(kind='bar', title='Price Variation by Brand (Standard Deviation)')
plt.ylabel('Price Std Dev')
plt.tight_layout()
plt.show()
```

```python
#.3 What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare.
sns.boxplot(x='Transmission', y='Price ($)', data=df)
plt.title('Car Price Distribution by Transmission Type')
plt.show()
```

```python
#4. What is the distribution of car prices across different regions?
sns.boxplot(x='Dealer_Region', y='Price ($)', data=df)
plt.title('Car Price by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5.  What is the distribution of cars based on body styles?
df['Body Style'].value_counts().plot(kind='bar', title='Car Count by Body Style')
plt.ylabel('Number of Cars')
plt.show()
```

```python
# 6.  How does the average selling price of cars vary by customer gender and annual income?
sns.boxplot(x='Gender', y='Price ($)', hue='Gender', data=df)
plt.title('Car Prices by Gender')
plt.show()

sns.scatterplot(x='Annual Income', y='Price ($)', hue='Gender', data=df)
plt.title('Price vs Income by Gender')
plt.show()
```

```python
#7.  What is the distribution of car prices by region, and how does the number of cars sold vary by region?
region_price = df.groupby('Dealer_Region')['Price ($)'].mean()
region_count = df['Dealer_Region'].value_counts()

fig, ax1 = plt.subplots()
region_price.plot(kind='bar', color='Black', ax=ax1, position=0, width=0.4, label='Avg Price')
ax2 = ax1.twinx()
region_count.plot(kind='bar', color='Red', ax=ax2, position=1, width=0.4, label='Car Count')
plt.title('Avg Price and Car Count by Region')
plt.show()
```

```python
# 8.  How does the average car price differ between cars with different engine sizes?
sns.lineplot(x='Engine', y='Price ($)', data=df)
plt.title('Avg Car Price by Engine Size')
plt.show()
```

```python
#9.  How do car prices vary based on the customerâ€™s annual income bracket?
df['IncomeBracket'] = pd.cut(df['Annual Income'], bins=5)
sns.boxplot(x='IncomeBracket', y='Price ($)', data=df)
plt.xticks(rotation=45)
plt.title('Car Price by Income Bracket')
plt.show()
```

```python
#10.  What are the top 5 car models with the highest number of sales, and how does their price distribution look?
top_models = df['Model'].value_counts().nlargest(5).index
sns.boxplot(x='Model', y='Price ($)', data=df[df['Model'].isin(top_models)])
plt.title('Price Distribution of Top 5 Selling Models')
plt.show()
```

```python
#11. How does car price vary with engine size across different car colors, and which colors have the highest price variation
color_variation = df.groupby('Color')['Price ($)'].std().sort_values(ascending=False)
print("Color with highest price variation:", color_variation.idxmax())
sns.scatterplot(x='Engine', y='Price ($)', hue='Color', data=df)
plt.title('Car Price vs Engine Size by Color')
plt.show()
```

```python
#12. Is there any seasonal trend in car sales based on the date of sale?
df['Date'] = pd.to_datetime(df['Date'])


df['Month'] = df['Date'].dt.month


monthly_sales = df['Month'].value_counts().sort_index()
sales_per_month = df['Month'].value_counts().sort_index()
sales_per_month.plot(kind='line', marker='o', title='Seasonal Sales Trend')
plt.xlabel('Month')
plt.ylabel('Cars Sold')
plt.show()
```

```python
# 13.  How does the car price distribution change when considering different combinations of body style and transmission type
sns.catplot(x='Body Style', y='Price ($)', hue='Transmission', kind='box', data=df, height=6, aspect=2)
plt.title('Car Price by Body Style and Transmission')
plt.xticks(rotation=45)
plt.show()
```

```python
#14. What is the correlation between car price, engine size, and annual income of customers, and how do these features interact
df = df[['Price ($)', 'Engine', 'Annual Income']]


correlation = df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix: Car Price ($), Engine , and Annual Income')
plt.show()


sns.pairplot(df, kind='reg', height=3, aspect=1.2)
plt.suptitle('Interaction Between Car Price ($), Engine , and Annual Income', y=1.02)
plt.show()
```

```python
#15.  How does the average car price vary across different car models and engine types?
if 'Engine' in df.columns:
    sns.catplot(x='Engine', y='Price ($)', kind='bar', data=df, height=6, aspect=2)
    plt.xticks(rotation=45)
    plt.title('Average Car Price by Model and Engine Type')
    plt.show()

if 'Model' in df.columns:
    sns.catplot(x='Model', y='Price ($)', kind='bar', data=df, height=6, aspect=2)
    plt.xticks(rotation=45)
    plt.title('Average Car Price by Model and Engine Type')
    plt.show()
```

```python
# 1.  What is the average rating for each product category?
avg_rating_by_category = df.groupby('category')['rating'].mean().sort_values(ascending=False)
print(avg_rating_by_category)
```

```python
#2. What are the top rating_count products by category/
top_rated_products = df.sort_values('rating_count', ascending=False).groupby('category').head(3)
print("Top Rating Count Products by Category:")
print(top_rated_products[['category', 'product_name', 'rating', 'rating_count']])
```

```python
# 3. What is the distribution of discounted prices vs. actual prices?
plt.figure(figsize=(10,5))
sns.histplot(df['actual_price'], color='red', label='Actual Price', kde=True)
sns.histplot(df['discounted_price'], color='green', label='Discounted Price', kde=True)
plt.legend()
plt.title("Price Distribution")
plt.show()
```

```python
# 4. How does the average discount percentage vary across categories?
avg_discount_by_category = df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)
print(avg_discount_by_category)
```

```python
# 5.  What are the most popular product names?
popular_product_names = df.groupby('product_name')['rating_count'].sum().sort_values(ascending=False).head(10)
print(popular_product_names)
```

```python
#6. What are the most popular product keywords?
from collections import Counter
import re

keywords = ' '.join(df['about_product'].dropna()).lower()
keywords = re.findall(r'\b[a-z]{4,}\b', keywords)  # 4+ letter words
common_keywords = Counter(keywords).most_common(10)
print(common_keywords)
```

```python
# 7.  What are the most popular product reviews?
reviews = ' '.join(df['review_content'].dropna()).lower()
review_words = re.findall(r'\b[a-z]{4,}\b', reviews)
common_reviews = Counter(review_words).most_common(10)
print(common_reviews)
```

```python
#8 What is the correlation between discounted_price and rating?
def clean_price(value):
    if isinstance(value, str):
        value = re.sub(r'[^\d.]', '', value)  # Remove â‚¹, commas, and symbols
    try:
        return float(value)
    except:
        return None

df['discounted_price'] = df['discounted_price'].apply(clean_price)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')



filtered_df = df.dropna(subset=['discounted_price', 'rating'])



correlation = filtered_df[['discounted_price', 'rating']].corr().iloc[0, 1]
print("Correlation between discounted price and rating:", correlation)
```

```python
#9. What are the Top 5 categories based on the highest ratings?
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Group by category and compute average rating
top_5_categories = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(5)

print("Top 5 Categories Based on Highest Average Ratings:")
print(top_5_categories)
```

```python
#10.  Identify any potential areas for improvement or optimization based on the data analysis.
median_price = df['actual_price'].median()
low_rating_high_price = df[(df['rating'] < 3.5) & (df['actual_price'] > median_price)]
print(low_rating_high_price[['product_name', 'category', 'rating', 'actual_price']].head())
```

```python
#2. What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=20, kde=True)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Number of Tracks')
plt.tight_layout()
plt.show()
```

```python
# 3.  Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Duration (ms)', y='Popularity')
plt.title('Popularity vs. Duration of Tracks')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.tight_layout()
plt.show()
```

```python
#4. Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot
df = pd.read_csv('spotify.csv')  # Replace with your path if needed

# Drop duplicates and nulls
df = df.dropna().drop_duplicates()

# Get artist track counts
artist_counts = df['Artist'].value_counts()

# Display the artist with the most tracks
top_artist = artist_counts.idxmax()
top_count = artist_counts.max()
print(f"Artist with the most tracks: {top_artist} ({top_count} tracks)")

# Plot countplot for top 10 artists by track count
top_10_artists = artist_counts.head(10)

plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['Artist'].isin(top_10_artists.index)],
              x='Artist', order=top_10_artists.index)
plt.title('Number of Tracks per Artist (Top 10)')
plt.xlabel('Artist')
plt.ylabel('Track Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#5. What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.
df = df.dropna().drop_duplicates()

# Sort the dataset by popularity (ascending) and get the top 5
least_popular_tracks = df.sort_values(by='Popularity', ascending=True).head(5)

# Display the artist name and track name
print("Top 5 Least Popular Tracks:")
print(least_popular_tracks[['Artist', 'Track Name', 'Popularity']])
```

```python
# 6.  Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist
df = df.dropna().drop_duplicates()


artist_avg_popularity = df.groupby('Artist')['Popularity'].mean()


top_5_avg_popular_artists = artist_avg_popularity.sort_values(ascending=False).head(5)


print("Top 5 Artists by Average Popularity:")
print(top_5_avg_popular_artists)
```

```python
# 7.  For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist
df = df.dropna().drop_duplicates()


top_5_artists = (
    df.groupby('Artist')['Popularity']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)


top_tracks = (
    df[df['Artist'].isin(top_5_artists.index)]
    .sort_values(by=['Artist', 'Popularity'], ascending=[True, False])
    .drop_duplicates(subset='Artist')
)


print("Most Popular Track for Each of the Top 5 Most Popular Artists:")
print(top_tracks[['Artist', 'Track Name', 'Popularity']])
```

```python
# 8. Visualize relationships between multiple numerical variables simultaneously using a pair plot?
df = df.dropna().drop_duplicates()


numerical_cols = ['Popularity', 'Duration (ms)']  # Add more if applicable

sns.pairplot(df[numerical_cols])
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
```

```python
#9.  Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot
df = pd.read_csv('spotify.csv')
df = df.dropna().drop_duplicates()


top_artists = df['Artist'].value_counts().head(10).index
filtered_df = df[df['Artist'].isin(top_artists)]


plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_df, x='Artist', y='Duration (ms)')
plt.title('Track Duration Variation Across Top 10 Artists (Box Plot)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
sns.violinplot(data=filtered_df, x='Artist', y='Duration (ms)', inner='box')
plt.title('Track Duration Distribution Across Top 10 Artists (Violin Plot)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
